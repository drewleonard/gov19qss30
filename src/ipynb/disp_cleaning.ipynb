```python
def get_groups(nsn):
    
    fsg = str(nsn[:2])
    if fsg not in fsg_group_map or fsg not in fsg_group_small_map:
        return "Unavailable", "Unavailable"
    return fsg_group_map[fsg], fsg_group_small_map[fsg]
    
```

```python
def get_county(lea):
    
    # Request geocode
    g = gmaps.geocode(lea)

    # Return if response is empty
    if not g:
        return "Unavailable", "Unavailable"
    
    # Find county
    county = None
    for e in g[0]['address_components']:
        if 'administrative_area_level_2' in e['types']:
            county = e
            break
        
    # Return if county is empty
    if not county:
        return "Unavailable", "Unavailable"
    
    # Get long and short names
    try:
        long_name = county['long_name']
        short_name = county['short_name']
    except:
        long_name = "Unavailable"
        short_name = "Unavailable"
    
    return long_name, short_name
    
```

```python
df_disp = pd.read_csv("./data/disp_merged_states.csv")
df_leso_codes = pd.read_csv("./data/leso_codes.csv")
```

```python
# Map leso codes to types
leso_map = {}
for i, e in df_leso_codes.iterrows():
    
    code = e['PSC CODE']
    name = e['PRODUCT AND SERVICE CODE NAME']
    
    if code not in leso_map:
        leso_map[code] = name
```

```python
# Attach leso names to df_disp
leso_names = []
for i, e in df_disp.iterrows():
    leso_code = get_leso_code(e['NSN'])
    leso_name = "Unavailable" if leso_code not in leso_map else leso_map[leso_code]
    leso_names.append(leso_name)
df_disp['leso_name'] = leso_names
```

```python
# Attach groups to df_disp
groups = []
groups_small = []
for i, e in df_disp.iterrows():
    nsn = e['NSN']
    group, group_small = get_groups(nsn)
    groups.append(group)
    groups_small.append(group_small)
df_disp['group'] = groups
df_disp['group_small'] = groups_small
```

```python
df_disp = df_disp[(df_disp['group'] != "Unavailable")]
df_disp = df_disp[(df_disp['group_small'] != "Unavailable")]
df_disp = df_disp.reset_index(drop = True)
```

```python
# Attach counties to df_disp
counties_map = {}

for i, e in df_disp.iterrows():
    
    if i % 5000 == 0:
        print("ROW: {}".format(i))
    
    # Get station address
    lea = e['Station Name (LEA)']
    state = e['State']
    address = "{}, {}".format(lea, state)
    
    if address in counties_map:
        continue
    
    county_long, county_short, = get_county(address)
    counties_map[address] = {
        'county_long': county_long,
        'county_short': county_short
    }
```

```python
counties_long = []
counties_short = []

for i, e in df_disp.iterrows():
    
    # Get station address
    lea = e['Station Name (LEA)']
    state = e['State']
    address = "{}, {}".format(lea, state)
    
    if address not in counties_map:
        counties_long.append('Unavailable')
        counties_short.append('Unavailable')
        continue
    
    county_long = counties_map[address]['county_long']
    county_short = counties_map[address]['county_short']
    
    counties_long.append(county_long)
    counties_short.append(county_short)

df_disp['county_long'] = counties_long
df_disp['county_short'] = counties_short
```
